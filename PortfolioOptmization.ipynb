```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
```

```python
from pypfopt.efficient_frontier import EfficientFrontier

def MarkowitzResults(start_year, start_month):
  df3 = df.query(f"Date.dt.year == {start_year} and Date.dt.month == {start_month}")

  listaSymbols = list(df.Symbol.unique())
  listaVolumes = []

  for i in listaSymbols:
    volume_local = (df3.query(f"Symbol == '{i}'")['Volume']).sum()
    listaVolumes.append([i, volume_local])

  lista_ordenada = sorted(listaVolumes, key=lambda x: x[1], reverse=True)[0:10]

  listaSimbolos = []
  for i in range(0, len(lista_ordenada)):
    listaSimbolos.append(lista_ordenada[i][0])

  df2 = pd.DataFrame()

  for i in range(0, len(listaSimbolos)):
    df1 = df3.query(f"Symbol == '{listaSimbolos[i]}'")
    df2 = pd.concat([df1, df2])

  df_port = pd.DataFrame()

  for symbol in listaSimbolos:
      df_symbol = df2.loc[df2["Symbol"] == symbol, ["Close"]]
      df_symbol = df_symbol.rename(columns={"Close": symbol})
      df_port = df_port.merge(df_symbol, how="outer", left_index=True, right_index=True)

  df_port = df_port[listaSimbolos]
  df_port

  from pypfopt.expected_returns import mean_historical_return
  from pypfopt.risk_models import CovarianceShrinkage

  mu = mean_historical_return(df_port)
  S = CovarianceShrinkage(df_port).ledoit_wolf()

  from pypfopt.efficient_frontier import EfficientFrontier

  ef = EfficientFrontier(mu, S)
  weights = ef.max_sharpe()

  cleaned_weights = ef.clean_weights()

  df3 = df.query(f"Date.dt.year == {start_year} and Date.dt.month == {start_month+1}")
  precos_iniciais = []
  for i in listaSimbolos:
    precos_iniciais.append(int(df3.query(f"Symbol == '{i}'")['Close'].iloc[0]))

  precos_finais = []
  for i in listaSimbolos:
    precos_finais.append(int(df3.query(f"Symbol == '{i}'")['Close'].iloc[len(df3.query(f"Symbol == '{i}'")['Close'])-1]))

    peso_normal = []
    for i in range(0, 10):
      peso_normal.append(0.1)

  peso_normal = pd.Series(peso_normal) 

  retorno_normal = (sum(precos_finais*peso_normal) - sum(precos_iniciais*peso_normal))/sum(precos_iniciais*peso_normal)

  j = 0
  soma_final = 0
  for i in listaSimbolos:
    soma_final += cleaned_weights[i]*precos_finais[j]
    j += 1

  j1 = 0
  soma_inicial = 0
  for i in listaSimbolos:
    soma_inicial += cleaned_weights[i]*precos_iniciais[j1]
    j1 += 1

  retorno_markowitz = (soma_final - soma_inicial)/soma_inicial

  return retorno_markowitz
```

```python
import matplotlib.pyplot as plt

normal_list = []
markowitz_list = []
indice = []

for i in range(2004,2022):
  for j in range(1, 12):
    try:
      resultado = MarkowitzResults(i,j)
      markowitz_list.append(resultado*100)
      indice.append(str(i)+'-'+str(j))
    except:
      print(".")
```

```python
import plotly.graph_objs as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=indice, y=markowitz_list, mode='lines', name='Markowitz'))

fig.update_layout(title='Comparação entre Normal e Markowitz', xaxis_title='Índice', yaxis_title='Valores', legend=dict(x=0, y=1))

fig.show()
```
